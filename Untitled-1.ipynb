```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
color= sns.color_palette()
```

```python
df = pd.read_csv('C:\\Users\\canbe\\OneDrive\\Masaüstü\\codes\\nba\\nba_players_shooting.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df['RANGE_num'] = np.sqrt(df.X**2 + df.Y**2)
df[['RANGE', 'RANGE_num']].sample(10)
```

```python
df.replace(['MADE', 'MISSED'], [1, 0]).groupby('SHOOTER').SCORE.mean().sort_values().plot.barh()
import matplotlib.ticker as mtick
plt.gca().xaxis.set_major_formatter(mtick.PercentFormatter(1))
plt.title("% of field goals")
```

```python
fig, ax = plt.subplots(1, 4, figsize=(19,4), sharey=True, sharex=True)
for i, shooter in enumerate(df.SHOOTER.unique()):
    data = df.query("SHOOTER == @shooter & SCORE == 'MADE'")
    sns.scatterplot(data=data, x='X', y='Y', ax= ax[i])
    ax[i].set_title(shooter)
```

```python
df['RANGE_region'] = np.select([df.RANGE_num > 23.9, df.RANGE_num > 9,], ['Beyond 3p Line', 'Within 3p Line'], default = 'Under basket')

df['RANGE_region'] = df['RANGE_region'].astype("category").cat.reorder_categories(['Under basket', 'Within 3p Line', 'Beyond 3p Line'], ordered = True)

sns.scatterplot(x = df.X, y = df.Y, hue = df.RANGE_region, legend = False, palette= 'pastel')
plt.title('The field, seperated into three areas')

df.sample(10)
```

```python
df.replace(['MADE', 'MISSED'], [1, 0]).groupby(['SHOOTER', 'RANGE_region']).SCORE.mean().unstack().pipe(sns.heatmap, cmap='Reds', square=True, annot=True, linecolor= 'Red', linewidths = 1, fmt = '.1%' )
plt.xlabel(None)
plt.ylabel(None)
plt.title("FG Ratio by Region")
```

```python
df.replace(['MADE', 'MISSED'], [0, 1]).groupby('DEFENDER').SCORE.mean().sort_values().plot.barh()

plt.gca().xaxis.set_major_formatter(mtick.PercentFormatter(1))
plt.title("% of missed shots made against the defender")
```

```python
df.replace(['MADE', 'MISSED'], [1, 0]).groupby(['SHOOTER', 'DEFENDER']).SCORE.mean().unstack().style.background_gradient(axis=1).format('{:,.1%}')
```
